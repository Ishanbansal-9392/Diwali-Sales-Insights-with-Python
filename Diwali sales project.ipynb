```python
import pandas as pd
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"E:\thetha academy\Python project\Diwali Sales Data.csv", encoding="unicode_escape")
df
```

```python
print(df.isna().sum())
```

```python
df=df.drop(columns="Status")
df
```

```python
print(df.isna().sum())
```

```python
df=df.drop(columns="unnamed1")
df
```

```python
print(df.isna().sum())
```

```python
df=df.dropna(how="any")
df=df.reset_index(drop=True)
df
```

```python
print(df.isna().sum())
```

```python
gender_count=df['Gender'].value_counts()
plt.bar(gender_count.index,gender_count.values)
for i in range(len(gender_count)):
    plt.text(gender_count.index[i],gender_count.values[i],gender_count.values[i],fontsize=10,color="green",ha="center",va="bottom")
plt.title('Gender Count')
plt.xlabel("Gender")
plt.ylabel("count")
plt.show()
```

```python
grouped=df.groupby(["Age Group","Gender"]).size().unstack(fill_value=0)
age_groups=list(grouped.index)
male_counts=list(grouped["M"])
female_counts=list(grouped["F"])
x=list(range(len(age_groups)))
y=[i+0.1 for i in x]
plt.bar(x,male_counts,width=0.2,color="red",label="Male")
for a in range(len(male_counts)):
    plt.text(x[a],male_counts[a],male_counts[a],fontsize=10,color="black",ha="center",va="bottom")
plt.bar(y,female_counts,width=0.2,color="green",label="Female")
for b in range(len(female_counts)):
    plt.text(x[b],female_counts[b],female_counts[b],fontsize=10,color="black",ha="center",va="bottom")
x_center=[(i+j)/2 for i,j in zip(x,y)]
plt.xticks(x_center,age_groups,)
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.title("Count of Gender by Age Group")
plt.legend()
plt.show()
```

```python
Marital_Status_count=df["Marital_Status"].value_counts().sort_index()
plt.bar(Marital_Status_count.index,Marital_Status_count.values)
for i in range(len(Marital_Status_count)):
    plt.text(Marital_Status_count.index[i],Marital_Status_count.values[i],Marital_Status_count.values[i],fontsize=10,color="green",ha="center",va="bottom")
plt.xticks([0, 1])
plt.title("Count vs Marital Status")
plt.xlabel("MaritalStatus")
plt.ylabel("count")
plt.show()
```

```python
top10_states=df["State"].value_counts().head(10)
plt.bar(top10_states.index, top10_states.values,)
for i in range(len(top10_states)):
    plt.text(i,top10_states.values[i], str(top10_states.values[i]),fontsize=10,color="green",ha="center",va="bottom")
plt.title("Top 10 States by Order Count")
plt.xticks(rotation=45)
plt.xlabel("State")
plt.ylabel("Order Count")
plt.show()
```

```python
sales_state=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
plt.bar(sales_state["State"],sales_state["Amount"])
for i in range(len(sales_state)):
    plt.text(i,sales_state["Amount"].iloc[i], round(sales_state["Amount"].iloc[i], 2),fontsize=10,color="green",ha="center",va="bottom",rotation=90)
plt.title("Top 10 States by Total Sales")
plt.xlabel("State")
plt.ylabel("Total Sales Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
occupation_count=df["Occupation"].value_counts()
plt.bar(occupation_count.index,occupation_count.values,)
for i in range(len(occupation_count)):
    plt.text(i,occupation_count.values[i], str(occupation_count.values[i]),fontsize=10,color="green",ha="center",va="bottom")
plt.title("Occupatios by Order Count")
plt.xticks(rotation=45)
plt.xlabel("Occupation")
plt.ylabel("Order Count")
plt.show()
```

```python
sales_occupation=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
plt.bar(sales_occupation["Occupation"],sales_occupation["Amount"])
for i in range(len(sales_occupation)):
    plt.text(i,sales_occupation["Amount"].iloc[i], round(sales_occupation["Amount"].iloc[i], 2),fontsize=10,color="green",ha="center",va="bottom",rotation=90)
plt.title("Occupations by Total Sales")
plt.xlabel("Occupation")
plt.ylabel("Total Sales Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
product_Category_count=df["Product_Category"].value_counts()
plt.bar(product_Category_count.index,product_Category_count.values,)
for i in range(len(product_Category_count)):
    plt.text(i,product_Category_count.values[i],str(product_Category_count.values[i]),fontsize=10,color="green",ha="center",va="bottom")
plt.title("Product_Category by Order Count")
plt.xticks(rotation=45)
plt.xlabel("Product_Category")
plt.ylabel("Order Count")
plt.show()
```

```python
sales_product_Category=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
plt.bar(sales_product_Category["Product_Category"],sales_product_Category["Amount"])
for i in range(len(sales_product_Category)):
    plt.text(i,sales_product_Category["Amount"].iloc[i],round(sales_product_Category["Amount"].iloc[i], 2),fontsize=10,color="green",ha="center",va="bottom",rotation=90)
plt.title("Product_Category by Total Sales")
plt.xlabel("Product_Category")
plt.ylabel("Total Sales Amount")
plt.xticks(rotation=45)
plt.show()
```
